#### 5. Check and deal with NaN values.

```python
#let's see first if we have any duplicated rows with nan values in those rows aka no info in that entry
duplicate_rows = data[data.duplicated()]
duplicate_rows
```

```python
data.columns
```

```python
data["customer"].isna().sum()
```

```python
data.isna().sum()
#we can see that we have NaN values in:
#state, response, months_since_last_claim, number_of_open_complaints, vehicle_class, vehicle_size, vehicle_type. 
```

```python
data["state"] = data["state"].fillna("Unknown")
data["response"] = data["response"].fillna("Unknown")
data["vehicle_class"] = data["vehicle_class"].fillna("Unknown")
data["vehicle_size"] = data["vehicle_size"].fillna("Unknown")
data["vehicle_type"] = data["vehicle_type"].fillna("Unknown")
data["months_since_last_claim"] = data["months_since_last_claim"].fillna(0)
data["number_of_open_complaints"] = data["number_of_open_complaints"].fillna(0)
data.isna().sum()
```

#### 6. Datetime format - Extract the months from the dataset and store in a separate column. Then filter the data to show only the information for the first quarter , ie. January, February and March. Hint: If data from March does not exist, consider only January and February. 

```python
data["effective_to_date"] = pd.to_datetime(data["effective_to_date"], errors='coerce')
```

```python
january_rows = data[data["effective_to_date"].dt.month == 1]
february_rows = data[data["effective_to_date"].dt.month == 2]
march_rows = data[data["effective_to_date"].dt.month == 3]
first_quarter_rows = pd.concat([january_rows,february_rows,march_rows])
first_quarter_rows
```
